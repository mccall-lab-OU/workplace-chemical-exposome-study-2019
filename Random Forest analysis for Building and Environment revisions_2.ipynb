```r
#example PERMANOVA code
#load libraries and files
library(vegan)
metadata <- read.csv('CHEM5170metadata_forAdonis.csv', header=TRUE, sep = ',', comment.char = "")
bcdm<-read.table('canberra_5xBlank_TICNorm_samplesOnly_noTeaching.txt', h=T,row.names=1,check=F)

#cut down distance matrix and metadata to just those samples that are in both 
rownames(metadata)<-metadata$Sample_ID
ix <- intersect(metadata[,1:1],rownames(bcdm))
metadata2<- metadata[ metadata$SampleID %in% ix, ]
dist <- bcdm[ix,ix]
dist <- as.matrix(dist)
dist <- as.dist(dist)

#run adonis (permanova) on the full dataset
msfull.adonis<-adonis(dist~metadata2$BuildingType, data=metadata2[labels(dist),])
msfull.adonis

#R2 above 10% indicates good/acceptable clustering; p-values can easily be influenced by differences in spread of the points, even if the centroids of each point cloud are the same
```
